Summary statistics such as mean, standard deviatation, and percentiles will not (I believe) do us little service for the purpose of our analysis due to the, once again, categorical nature of our variables.

**Methodology:**
To determine if there exists a correlation between sex life and loneliness, we will perform a Fisher Exact Test on contingency tables between **CONNECTION_activities_sex_p3m** and all the previously mentioned **LONELY** variables. This method will have us generating simulated contingency samples under a null hypothesis, after which we will use a Monte Carlo simulation (I'm fighting chatgpt to make the simulations and the contingency tables work) to calculate our p-value. For the purposes of hypothesis testing, our **$\alpha = 0.05$**, which will be compared to our p-value. For the purposes of the test, we will assume that our observations are independent of each other and the total of our columns and rows are fixed. There are of course other variables that we are missing out on due to the nature of the question (*e.g. people with partners are most likely more sexually active and simultaneously less lonely than those without a partner. But that is not the focus of the question.*)

**Hypothesis:**

$H_0$: **There is no correlation between sex life and (whatever loneliness variable)**

$H_1$: **A correlation does exist between sex life and (whatever loneliness variable)**

**Possible results?** Two possibilities exist for each loneliness variable: Either $p <= \alpha$ (We reject $H_0$) or $p >= \alpha$ (We fail to reject $H_0$).

**Question relevance?** With these results we may determine whether or not a person's sexual activity relates to their personal state of loneliness.
