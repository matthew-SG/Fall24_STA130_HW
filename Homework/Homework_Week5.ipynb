**Question 1:**

The key factor that makes the difference between ideas that can, and cannot be examined and tested statistically is the existence of data behind the ideas in question. An example used in the video was how you cannot examine and statistically test whether or not Hillary Clinton or Donald Trump would be a better president, because there's no data of a Clinton administration that can be compared with a Trump administration. 

The key criteria defining what a good null hypothesis is **A:** there is data behind the topic of the hypothesis and **B:** the hypothesis can actually be tested with the present data. 

The difference between a null hypothesis and an alternative hypothesis in the context of hypothesis testing is that the null hypothesis is a statement/idea that you are trying to disprove, while the alternative hypothesis is the assumption you can favour over the null hypothesis IF it can be rejected. You go into hypothesis testing assuming the null hypothesis is correct, and see if an examination of present data can contradict with that initial assumption.

**Question 2:**

What that sentence means is that when you're hypothesis testing, that being checking if the null hypothesis (the statement we're trying to disprove) can be rejected or not (statistical observations imply that the statement is confidently false), you're testing your hypothesis in relation the parameters of a population (the attributes of a theoretical dataset that contains every possible data-point for some observation). Despite the fact that you are using sample values (values that represent a small part of a population) to calculate sample statistics such as sample mean (the average of sample values) which you then use for your hypothesis testing, any conclusion you come to will refer to the related population parameter, in the case of sample mean that would be the population mean (average of that theoretical data-set). For example, if you cannot reject a null hypothesis of a population mean, that means you can assume the population mean to be the value stated in your null hypothesis (symbolized as 𝜇0). If can reject it, than you can assume it cannot.

**Question 5:**

We want to see whether or not Gunturkun's observation of couples tendency to tilt their heads to right when kissing is accurate for the population. So, under the null hypothesis that couples don't have left or right tendencies to kiss (0.5 proportion), we will find the p-value to determine how probable the null hypothesis is:

````python
import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(58)  # You can use any integer here

# Load the dataset
url = 'https://raw.githubusercontent.com/matthew-SG/Fall24_STA130_HW/refs/heads/main/Datasets/week_5_kisser.csv'
data = pd.read_csv(url)

# Observed data
observed_left = data[data['Head Tilt'] == 'Left']['Count'].values[0]
observed_right = data[data['Head Tilt'] == 'Right']['Count'].values[0]
n_total = observed_left + observed_right

# Calculate the observed proportion of right head tilt
observed_proportion_right = observed_right / n_total

# Null hypothesis: Proportion of left and right head tilt is equal (0.5)
population_parameter_value_under_H0 = 0.5

# Number of simulations
number_of_simulations = 10000

# Simulate under the null hypothesis
simulated_right_counts = np.random.binomial(n_total, population_parameter_value_under_H0, number_of_simulations)
simulated_test_statistics = simulated_right_counts / n_total  # Proportion of right head tilts

# Calculate the observed test statistic
observed_test_statistic = observed_proportion_right

# Calculate if simulated stats are as or more extreme than the observed
SimTestStats_as_or_more_extreme_than_ObsTestStat = \
    abs(simulated_test_statistics - population_parameter_value_under_H0) >= \
    abs(observed_test_statistic - population_parameter_value_under_H0)

# Calculate the p-value
p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations

# Print the results
print(f"Observed proportion of right head tilt: {observed_proportion_right:.4f}")
print(f"P-value: {p_value:.4f}")
````

As we can see, when comparing the simulated statistics of the null hypothesis and the observed statistics of Gunturkun's sample, the p-value is significantly low (0.0012), meaning we have strong evidence against the null hypothesis, meaning that its ridiculous/unprobable for Gunturkun's observation to occur under the assumption that the population has no head tilt tendency while kissing.

````python
# Calculate the p-value for a one-tailed (right-tailed) hypothesis test
# How many bootstrapped statistics generated under H0 are "as or more extreme"
# (i.e., greater than or equal to the observed statistic, relative to H0)?

observed_test_statistic = (patient_data.HealthScoreChange > 0).mean()
simulated_test_statistics = IncreaseProportionSimulations_underH0random

# For a right-tailed test: Only consider simulated statistics greater than or equal to the observed
SimTestStats_as_or_more_extreme_than_ObsTestStat = \
    simulated_test_statistics >= observed_test_statistic

# Calculate the p-value
p_value = (SimTestStats_as_or_more_extreme_than_ObsTestStat).sum() / number_of_simulations

# Output the results
print("Number of Simulations: ", number_of_simulations, "\n\n",
      "Number of simulated test statistics (under HO)\n",
      'that are "as or more extreme" than the observed test statistic: ',
      SimTestStats_as_or_more_extreme_than_ObsTestStat.sum(), "\n\n",
      'p-value\n(= simulations "as or more extreme" / total simulations): ', p_value, sep="")
````

**Changes Made?**

Since we're doing a one tailed test, we only care if simulated test statistics are as great or greater than (right tailed) or as less or less than (left tailed) the observed test statistic. So the changes that were made to the original computation code was that the absolute function was dropped, as we were no longer examining whether or not the symmetrical distance between the simulated test statistics and the null hypothesis parameter were as or more extreme than the distance between the observed test statistic and the same parameter.

As we were doing a right tailed test (as the observed test statistic is greater than the null hypothesis parameter), we were only checking for and counting the test statistics that were equal to or greater than the observed test statistic. This changes the interpretation of the hypothesis test, because instead of determining the probability that the test statistics deviate away from a parameter beyond a value in both directions, a one tailed test determines the probability of a deviation in a singular direction. We should expect the p-value to be smaller in a one tailed analysis **IF** our observation test statistic deviates well into the chosen direction of the analysis. For example, if you were to conduct a left tailed test with the vaccine dataset, the p-value would be **larger** since a vast majority of the simulated test statistics are less than the observed test statistics. The dataset suggests that health improves from the vaccine (skews right), while a left tailed analysis would, in this scenario, be used to check if the observed test statistic is unusually less than the null hypothesis (which it obviously not).

**Introduction:**
In the 1920s, Ronald Fisher and his colleague Dr. Muriel Bristol carried out a test off a pure curiousity: could you taste in your tea whether the milk or the tea were poured first? From Fisher's experiment with Bristol, out of 8 cups that were drank by Bristol, she guessed all 8 correctly. Since 8 is a small sample size, and one person is of course one person, our test is the same test but replicated tenfold. We have 80 students, all who which drank a singular cup of tea, and out of the 80 students who drank the tea, 49 of them were able to correctly identify whether the milk or tea came first.

In this problem, we're going to assume the null hypothesis that it is impossible to taste the difference in the order of liquid poured, that it is random chance. So if you have a random chance of guessing whether or not one of two liquids were poured first, then you would have a 50% chance of guessing correctly, which will be our null hypothesis, that you are to guess it correctly half of the time. Our alternative hypothesis will be that you **can** taste the difference in order, and that you would be able to guess correctly more often than not. Let's set our alpha significance level to be 0.05, so we may know how confident we can be in our p-value.

**Analysis:**

First, lets examine a basic histogram of our values:

````python
import pandas as pd
import plotly.express as px

# Load the dataset
url = 'https://raw.githubusercontent.com/matthew-SG/Fall24_STA130_HW/refs/heads/main/Datasets/milk_tasting.csv'
data = pd.read_csv(url)

# Reshape the data for better plotting (long format)
melted_data = pd.melt(data, var_name='Guess', value_name='Count')

# Create a bar plot
fig = px.bar(melted_data, x='Guess', y='Count', title='Tea Tasting: Guessed Correctly vs Guessed Incorrectly')

# Show the plot
fig.show(renderer='png')

# Extract the counts for "Guessed Correctly" and "Guessed Incorrectly"
guessed_correctly = data['Guessed Correctly'].values[0]
guessed_incorrectly = data['Guessed Incorrectly'].values[0]

# Calculate the total guesses
total_guesses = guessed_correctly + guessed_incorrectly

# Calculate the proportion of "Guessed Correctly"
proportion_correct = guessed_correctly / total_guesses

# Print the result
print(f"Proportion of Guessed Correctly: {proportion_correct:.4f}")
````

We can see that in our observation, approximately 61% of the guesses are correct.

Now, let's examine the proportions of the simulated proportions under the null hypothesis.

````python
import pandas as pd
import numpy as np
import plotly.express as px

# Set a random seed for reproducibility
np.random.seed(42)

# Load the dataset
url = 'https://raw.githubusercontent.com/matthew-SG/Fall24_STA130_HW/refs/heads/main/Datasets/milk_tasting.csv'
data = pd.read_csv(url)

# Observed data
guessed_correctly = data['Guessed Correctly'].values[0]
guessed_incorrectly = data['Guessed Incorrectly'].values[0]
total_guesses = guessed_correctly + guessed_incorrectly

# Null hypothesis: Probability of guessing correctly is 50% (0.5)
probability_correct_under_H0 = 0.5

# Number of simulations (bootstrapping)
num_simulations = 10000

# Simulate the bootstrapped dataset using binomial distribution
# np.random.binomial(n, p, size) generates `size` samples of `n` trials with success probability `p`
simulated_correct_guesses = np.random.binomial(total_guesses, probability_correct_under_H0, num_simulations)

# Calculate the proportion of correct guesses in each simulation
simulated_proportions = simulated_correct_guesses / total_guesses

# Visualize the simulated proportions using a histogram
fig = px.histogram(simulated_proportions, nbins=30, title="Bootstrapped Proportions of Correct Guesses (Under H0: p=0.5)",
                   labels={'value':'Proportion of Guessed Correctly', 'count':'Frequency'})

# Show the plot
fig.show(renderer='png')
````

Lets just overlay a KDE over the bootstrapped proportions so we may examine the distribution:

````python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.stats import gaussian_kde

# Set a random seed for reproducibility
np.random.seed(42)

# Load the dataset
url = 'https://raw.githubusercontent.com/matthew-SG/Fall24_STA130_HW/refs/heads/main/Datasets/milk_tasting.csv'
data = pd.read_csv(url)

# Observed data
guessed_correctly = data['Guessed Correctly'].values[0]
guessed_incorrectly = data['Guessed Incorrectly'].values[0]
total_guesses = guessed_correctly + guessed_incorrectly

# Null hypothesis: Probability of guessing correctly is 50% (0.5)
probability_correct_under_H0 = 0.5

# Number of simulations (bootstrapping)
num_simulations = 10000

# Simulate the bootstrapped dataset using binomial distribution
simulated_correct_guesses = np.random.binomial(total_guesses, probability_correct_under_H0, num_simulations)

# Calculate the proportion of correct guesses in each simulation
simulated_proportions = simulated_correct_guesses / total_guesses

# Create a histogram using Plotly
fig = go.Figure()

# Add histogram with darker color
fig.add_trace(go.Histogram(x=simulated_proportions, nbinsx=30, name='Simulated Proportions', 
                             opacity=0.7, histnorm='probability density', 
                             marker=dict(color='blue')))

# Create KDE using scipy
kde = gaussian_kde(simulated_proportions)
x_vals = np.linspace(0, 1, 100)  # Range from 0 to 1
kde_vals = kde(x_vals)

# Add the KDE line
fig.add_trace(go.Scatter(x=x_vals, y=kde_vals, mode='lines', name='KDE', 
                         line=dict(color='orange', width=2)))

# Update layout for better visualization
fig.update_layout(
    title="Bootstrapped Proportions of Correct Guesses with KDE (Under H0: p=0.5)",
    xaxis_title="Proportion of Guessed Correctly",
    yaxis_title="Density",
    showlegend=True
)

# Show the plot
fig.show(renderer='png')
````

Now that we have the simulation of the null hypothesis, let's calculate the p-value using a right tailed test, checking how rare it is for the proportions under the null hypothesis to be as greatly extreme as the observation proportions:

````python
# Assuming you have already run the previous code and have the necessary variables defined

# Step 1: Calculate the observed proportion of correct guesses
observed_proportion = guessed_correctly / total_guesses

# Step 2: Calculate the right-tailed p-value
# Count how many of the simulated proportions are greater than or equal to the observed proportion
p_value = np.mean(simulated_proportions >= observed_proportion)

# Print the results
print(f"Observed Proportion of Guessed Correctly: {observed_proportion:.4f}")
print(f"Right-tailed P-value: {p_value:.4f}")
````

**Conclusion:**

Since our right tailed p-value is **less** than our established alpha significance level (0.05), we can be confident that the null hypothesis, that you cannot taste the order in which the tea is poured, is false. Through that conclusion, we can assume that the alternative hypothesis is true, that you **can** taste the difference in the order at which the milk and tea is poured.

### Summary of P-Value Calculation for Right-Tailed Test

1. **Objective**: Determine the right-tailed p-value comparing the observed proportion of correct guesses against simulated proportions under the null hypothesis.

2. **Definitions**:
   - **Observed Proportion**: The ratio of correctly guessed items to total guesses.
   - **Simulated Proportions**: The proportions obtained from bootstrapping the dataset, assuming the null hypothesis that the guessing probability is 50%.

3. **Steps**:
   - **Calculate the Observed Proportion**:
     \[
     \text{Observed Proportion} = \frac{\text{Guessed Correctly}}{\text{Total Guesses}}
     \]
   - **Count Simulated Proportions**:
     Count how many of the simulated proportions are greater than or equal to the observed proportion.
   - **Calculate the P-Value**:
     \[
     \text{P-Value} = \frac{\text{Count of Simulated Proportions} \geq \text{Observed Proportion}}{\text{Total Simulations}}
     \]

4. **Output**: 
   - Print the observed proportion and the right-tailed p-value.
