**1.1**

**Simple Linear Regression:**

- Uses one predictor variable to predict the outcome variable
- Focus is on the characterization of the (linear) relationship between a singular predictor and outcome, if there exists any
- Interpretation is based on the slope coefficient, $\beta_1$
- Represented by the model $Y_i = \beta_0 + \beta_1 x_i + \epsilon_i$, where $Y_i$ is our outcome, $\beta_0$ is the intercept, $\beta_1 x_i$ is the predictor variable and its slope coefficient, and $\epsilon_i$ is the error term.

**Multiple Linear Regression:**

- Utilizes two or more predictor variables to predict the outcome variable
- Focus is on influence/relationship that multiple predictor variables together have on the outcome variable
- Interpretation is based on the consideration of the coefficient for each predictor ($\beta_1, \beta_2,$ etc.), typically assuming that other predictor variables are constant so that the relationship between a single predictor and the outcome may be assessed with the consideration of the other variables.
- Represented by the model $Y_i = \beta_0 + \beta_1 x_{1i} + \beta_2 x_{2i} + \cdots + \epsilon_i$, where $Y_i$ is our outcome, $\beta_0$ is the intercept, $\beta_1 x_{1i}, \beta_2 x_{2i}, \text{and} \cdots$ are the predictor variables and their respective slope coefficients (with $\cdots$ being any subsequent predictors), and $\epsilon_i$ is the error term.

The benefits that multiple linear regression has over simple linear regression is that multiple linear regression accounts for the multiple influences/variables that affect the outcome, rather than the limited scope of one predictor variable for simple linear regression. Multiple linear regression also allows for you to analyze the individual effect of a predictor variable on the outcome, allowing you to see its isolated contribution regardless of any other factors. Multiple linear regression is also more versatile, as it can accomodate terms, such as categorical predictor variables or interactions between two predictors. Interaction terms are one of the key differences between simple and multiple linear regression, as allowing predictor variables to depend on each other in the model allows for more nuance in the interpretation of the outcome.

**1.2**

A continuous variable is a numeric variable that is able to take on any value within a given range. This can range from something like time, weight, height, distance, etc. A indicator variable is a binary variable that represents the presence of a characteristic or category; this is how categorical data is implemented into a simple linear regression model. The difference between these linear forms is that for continuous variables:
- $Y_i = \beta_0 + \beta_1 x_i + \epsilon_i$
    - $Y_i$: Outcome
    - $\beta_0$: Intercept Coefficient
    - $\beta_1$: Slope Coefficient
    - $x_i$: Predictor Variable
    - $\epsilon_i$: Error Term
- $\beta_0$ would represent how much $Y_i$ would change for a one unit change in $x_i$ while $\beta_0$ represents the average value of $Y_i$ when $x_i$ is zero.

For indicator variables:
- $Y_i = \beta_0 + \beta_1 1_{[x_i = \text{"B"}]}(x_i) + \epsilon_i$
    - $Y_i$: The outcome
    - $\beta_0$: Intercept
    - $\beta_1$: Contrast Coefficient
    - $1_{[x_i = \text{"B"}]}(x_i)$: Indicator Variable
    - $\epsilon_i$: Error Term
- $\beta_0$ Represents the mean of the reference group (Called Category "A"), $1_{[x_i = \text{"B"}]}(x_i)$ takes on a value of 1 if $x_i$ is a member of Category "B", and takes a value of 0 otherwise (why it's a binary variable), and $\beta_1$ represents the difference in means between "A" and "B" (the difference in average between "A" and "B").

A continuous variable models the continuous relationship between the outcome and the predictor, whereas the indicator variable models the difference in average outcome between two categories.

**1.3**

When you introduce a single indicator variable alongside a continuous variable you get the model:
- $ \text{outcome} = \beta_0 + \beta_1 \text{predictor} + \beta_2 \text{1(indicator)} $

where $\beta_2$ represents the difference in intercepts the between members and non-members of the indicator's specified category. What this effectively does is create two parallel lines with the same slope, with each line representing two different groups. The purpose of this is to model the average difference between the outcomes of two groups, under the assumption that the relationship between the predictor and the outcome is the same for both groups. While simple linear regression only allows for the analysis of the relationship between a predictor and an outcome, this form of multiple linear regression allows for the analysis of the average effect that two distinct groups have in relation to the outcome.

**1.4**

The linear form of adding an interaction between a continuous and an indicator variable is the following:
- $ \text{outcome} = \beta_0 + \beta_1 \text{predictor} + \beta_2 \text{1(indicator)} + \beta_3 \text{predictor} \times \text{1(indicator)} $

Where $\beta_3$ represents the **difference in slopes** between the two groups. What effect that this has is that instead of there being two parallel lines with the same slope, we instead introduce the possibility of getting two unique lines for two different groups. The inclusion of the interaction term makes it so that the group that a predictor belongs to affects the relationship it has to the outcome variable. We are no longer analyzing the average difference in outcomes of two groups, but rather difference of relationship to the outcome between two groups.

**1.5**

The linear form for a Multiple Linear Regression model based only on indicator variables dervied from a non-binary categorical variable is the following:
- $\text{outcome} = \beta_0 + \beta_1\text{1(category 2)} + \beta_2\text{2(category 3)} + \cdots + \beta_{n-1}\text{1(category k)}$

Where $\beta_0$ represents the average outcome for the **basline category** (the category that is not listed in the model), while $\beta_1, \beta_2 \cdots \beta_{k-1}$ are the coefficients for each indicator, representing the average difference in outcome between those categories and the baseline category. "k" in this context represents the number of categories within the categorical variable, with "category 1" being the baseline category.

When it comes to the interpretation of this model, the baseline category chosen affects value of the coefficients for the entire model, as the intercept is based on the **average outcome for the baseline category** while the coefficients for each indicator variable are the **average difference in the outcome between the baseline and the related category.**

For example, say you had a dataset that represents the exam scores of various students, with a continuous variable representing exam scores, exam_scores and a categorical variable, class_section, representing a student's section (call them section "A", "B", and "C"). Let's say that we want to see the effect that a student's section has on their exam score, and we choose section A to be our baseline category:
- Let's say we setup our model to be the following:
    - $\text{exam_score}=\beta_0+\beta_1\text{1(class_section = "B")}+\beta_2\text{1(class_section = "C")}$
- $\beta_0$ would represent the average exam scores of section A
- $\beta_1$ would represent the average difference in exam scores between section A and B
- $\beta_2$ would represent the average difference in exam scores between section A and C

- Now what if we change the baseline to be section B, and we change our model to be the following:
    - $\text{exam_score}=\beta_0+\beta_1\text{1(class_section = "A")}+\beta_2\text{1(class_section = "C")}$
- In the case, $\beta_0$ would represen the average exam scores of section B
- $\beta_1$ would represent the average difference in exam scores between section B and A
- $\beta_2$ would represent the avearge difference in exam scores between section B and C

Evidently, the values of the coefficient are dependent on the baseline category chosen, as the average exam scores for A could be different than B and the average difference in exam scores between A and C could differ from the average difference in exam scores between B and C.

When it comes to the overall structure of the model, the reason why we use k-1 indicator variables is to avoid **perfect multicollinearity**. If you were to use K number of indicator variables (every category has a related indicator variable), the sum of all of them for each unique observation would be 1. Due to this, you would be able to predict the value of each indicator variable based off the value of another, which makes it impossible to examine the effect that a unique category has on the overall data. By dropping an indicator variable, we avoid that linear dependency as the coefficients are now relative only to the intercept, meaning we would be able to isolate and estimate their individually unique effects.

The binary variable encoding implemented in the model has the purpose of isolating each unique category within the categorical variable for the purpose of analysis. By representing the categorical variable as a set of binary variables with each one corresponding to one unique category, we are able to signal which category an observation belongs to, allowing for the examination of the properties possessed by all the observations belonging to each group.  

**Question 2:**

The outcome variable ($S_i$) for this scenario would be the total sales of the comapny's sports equipment.

The predictor variables would be the the amount spent on TV advertisements ($t_i$), and the amount spent on online advertisements ($o_i$). 

An interaction that might need to be taken into account when predicting the outcome is the effect that the allocations of funds towards one medium or another have on each other. The effectiveness of the advertisement of television could be related to the amount spent on online advertisement and vice versa.

***Omitting interaction:***

$S_i = \beta_0 + \beta_tt_i + \beta_oo_i + \epsilon$


***Taking interaction into consideration:***

$S_i = \beta_0 + \beta_tt_i + \beta_oo_i + \beta_{to}(t_i \times o_i) + \epsilon$


**2.1**

The first model can be used to make outcome predictions under the assumption that the effects of advertisement spending for both mediums is **additive**. What this means is that their impact on the sale is the sum of their individual effects. $\beta_o \text{and} beta_t$ both represent the average change in sales in relation to their respective mediums while the other medium is held constant. Basically, we assume that the spending for both mediums do not influence each other.

The second model can be used to make outcome predictions under the assumption that the effects of advertisement spending for both mediums is **synergistic**. What this means is that their impact on sales depends on one another. Compared the last model, we assume under this model that the spending for both mediums influence each other.

The difference between models with and without interaction is that models **without** interaction (as mentioned earlier) assumes that each (predictor) variable possesss its own independent effect on the outcome, regardless of the other variables. Models **with** interaction assumes that the effect of each variable is dependent on the value/state of the variable it is interacting with. A one-unit increase in a variable may be larger or smaller based on the state of another variable.

**2.2**

In order to use the two formulas to make predictions of the outcome supposing that the two advertisement budgets are either "high" or "low", we have to represent both advertisement budgets as **indicator** variables based on if an observation for either budget has a spending of "high" or "low". 

Both types of models would either use 'low' or 'high' as the baseline category for each advertisement variable **individually**. For the purposes of explanations, I am using low as the baseline category.

The interpretation of the coefficients would of course change, with $\beta_0$ representing the average sales when both tv and online advertisement budgets are low, while $\beta_t$ and $\beta_o$ represent the average difference in sales when TV budget is high while online budget is low and vice-versa, respectively. For the interaction model, $\beta_to$ (or would be $\beta_3$) represents the additional average difference in sales that would be observed when both TV and Online advertisment budgets are 'high' compared to the sum of their individual 'high' differences. 

The interaction models would look like the following:

**Additive**

$S_i = \beta_0 + \beta_t1_{[t_i = \text{'high'}]} + \beta_o1_{[o_i = \text{'high'}]}(o_i)+ \epsilon$

**Synergistic**

$S_i = \beta_0 + \beta_t1_{[t_i = \text{'high'}]}(t_i) + \beta_o1_{[o_i = \text{'high'}]}(o_i) + 
\beta_{to}(1_{[t_i = \text{'high'}]}(t_i)\times1_{[o_i = \text{'high'}]}(o_i)) + \epsilon$

$R^2$ represents the proportion of variation explained by ***the model***. P-values that provide *strong* or *very strong evidence* against the *null hypothesis* only ***provides strong evidence*** that there is a linear relation between the predictor and the outcome. The coefficients represent how much the outcome will change for a one unit change in the related variable. 

With those three concepts in mind, the fact that $R^2$ has a low value only means that there are additional variables outside the model that influence the outcome.
```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

This cell separates the pokeaman dataset into two evenly subsets; one intended for fitting/training and the other for testing the trained model.

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

This cell defines a model prediciting the HP of a pokemon based on the additive effect of their attack and defense relative to their HP individually, based off the previously defined training subset. A line is then fitted to that model, and we are given a summary of the fitted model, with there being moderate evidence against no linear correlation for defense and HP, and very strong evidence against linear correlation for attack and HP.

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

The model first predicts further outcomes using the fitted model's equation. Variable y is then assigned the HP values from the test subset. The in-sample $R^2$ is then pulled from the fitted model, while the out-of-sample $R^2$ is calculated based on the values predicted from the fitted model. Comparing the two models shows to what extent the new model learned from the training data.

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

This cell creates a multiple linear regression model for pokemon HP based on the trained data, where every included variable interacts with each other. The model is then fit, and then a summary of the fitted model is displayed. Due to the complexity of the model, containing six predictor models, the model could very well be an overfit model. This is due to the fact that it could look too deeply into the patterns possessed by the indicators, introducing the potential of noise and fluctuations being captured by the model, obscuring the actual relations between the variables.

Due to that, this model could result in poor generalizability; it would be poor at predicting new values.

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

In this cell, we retrieve the in-sample $R^2$ from the model we constructed earlier, and then we calculate the out-of-sample $R^2$ from the fitted model and the test subset. As we can see, the in sample $R^2$ greatly deviates away from the out of sample $R^2$, showing that the model did overfit the training data.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

linear_form = 'HP ~ Attack + Defense'

in_sample_rsquared = []
out_of_sample_rsquared = []

for i in range(100): 
    msk = np.random.rand(len(pokeaman)) < 0.5 
    pokeaman_train = pokeaman[msk]
    pokeaman_test = pokeaman[~msk]

    # Fit the model on the training data
    model_spec = smf.ols(formula=linear_form, data=pokeaman_train)
    model_fit = model_spec.fit()

    # Calculate in-sample R-squared
    in_sample_rsquared.append(model_fit.rsquared)

    # Predict on the testing data
    predictions = model_fit.predict(pokeaman_test)

    # Align indices to avoid length mismatch
    predictions = predictions.reindex(pokeaman_test.index)

    # Calculate out-of-sample R-squared
    corr = np.corrcoef(pokeaman_test['HP'], predictions)[0, 1]
    out_of_sample_rsquared.append(corr**2)

# Combine the R-squared values into a DataFrame for Plotly Express
df_rsquared = pd.DataFrame({
    'R-squared': in_sample_rsquared + out_of_sample_rsquared,
    'Type': ['In-Sample'] * 100 + ['Out-of-Sample'] * 100
})

# Create the histogram using Plotly Express
fig = px.histogram(df_rsquared, x='R-squared', color='Type', barmode='overlay',
                   title='Distribution of In-Sample vs. Out-of-Sample R-squared',
                   labels={'R-squared': 'R-squared Value'},
                   opacity=0.7) 

fig.show(renderer='png')
```

What this results of this demonstration represents is the distribution of $R^2$ values, showing how much the performance of the models can fluctuate based on the data split. What we can determine from this visualization is that the fitted model does not generalize too well, as the distribution of the in sample $R^2$ deviates more to the right compared to its out of sample counterpart that is more concentrated on the left. As a hypothetical, if the in sample was more concentrated towards high values and less spread, that would imply that the fitted model is overfitting, and as extension would be generalizing poorly due to that deviation away from the out of sample $R^2$ values.
