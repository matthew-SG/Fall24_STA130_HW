The reason why there is a discrepancy between the amount of rows and columns given by df.shape and df.describe() is due to the difference in their functions. df.describe() only analyzes the numerical columns of a DataFrame for statistical purposes, hence why the non-numerical variables (age and fur-colour for example) are not present. df.shape is used to provide the dimensions of a DataFrame, so it counts every row and column. This is also why values in the 'count' column could be different, because the count column only reports non-null entries (rows). However, since the numerical observations of the squirrel dataset contains no null entries, each count is equal to the total rows of the dataset.

Question 5: The difference between an attribute and a method is the way that they interact with objects. Attributes directly access the data of an object without any additional computations, like df.shape, which directly pulls the dimensions of a DataFrame. Methods, indicated by parentheses, perform computations based on the data of an object in order to generate and return results that are not available or held within an object. df.describe() is an example of this, as it accesses a DataFrame and generates statistics such as mean and percentages that are not locally available within the DataFrame.
